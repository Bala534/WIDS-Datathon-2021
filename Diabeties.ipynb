```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import tensorflow as tf
import keras
```

```
data = pd.read_csv("TrainingWiDS2021.csv")
data.head()
```

```
data.shape
```

```
del data['Unnamed: 0']
del data['ethnicity']
del data['hospital_id']
del data['hospital_admit_source']
del data['icu_admit_source']
del data['icu_id']
del data['icu_stay_type']
del data['icu_type']
del data['pre_icu_los_days']
del data['readmission_status']
```

```
data.shape
```

```
data.columns
```

```
data.isna().any()
```

```
data['gender'] = data.gender.map(dict(M=1, F=0))
```

```
data =data.fillna(method="ffill")
```

```
data =data.fillna(method="bfill")
```

```
data.isna().any()
```

```
data.isnull().sum().sum()
```

```
data.describe()
```

```
data.describe(include="all")
```

```
data.info()
```

```
from sklearn.linear_model import LogisticRegression
```

```
x = data.loc[:,"encounter_id":"solid_tumor_with_metastasis"]
y = data['diabetes_mellitus']
```

```
model1 = LogisticRegression(max_iter=2000)
model1.fit(x,y)
```

```
y_train_predict1 = model1.predict(x)
```

```
print(pd.DataFrame({"Actual":data["diabetes_mellitus"],"Predicted":y_train_predict1}))
```

```
from sklearn.metrics import roc_auc_score
```

```
roc_auc_score(y,y_train_predict1)
```

```
from sklearn.naive_bayes import GaussianNB
```

```
model2 = GaussianNB()
model2.fit(x,y)
```

```
y_train_predict2 = model2.predict(x)
y_train_predict2
```

```
print(pd.DataFrame({"Actual":data["diabetes_mellitus"],"Predicted":y_train_predict2}))
```

```
roc_auc_score(y,y_train_predict2)
```

```
from sklearn.svm import SVC
```

```
model3 = SVC()
model3.fit(x,y)
```

```
y_train_predict3 = model3.predict(x)
print(y_train_predict3)
```

```
print(pd.DataFrame({"Actual":data["diabetes_mellitus"],"Predicted":y_train_predict3}))
```

```
roc_auc_score(y,y_train_predict3)
```
